```python
def plot_variable_pie(data_table,variable): 
    #准备绘图数据 
    data =data_table[variable].value_counts() 
    #创建饼图 
    fig1,ax1= plt.subplots() 
    ax1.pie(data,labels=data.index,autopct='%1.1f%%') 
    plt.axis('equal') 
    
```

```python
plot_variable_pie(df1,'父亲职业类别')
```

```python
plot_variable_pie(df1,'母亲职业类别')
```
